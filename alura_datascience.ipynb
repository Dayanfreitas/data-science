# Tipos de variáveis


```python
tmdb= pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.original_language.unique()
```

```python
# serie
tmdb.original_language.value_counts()
```

```python
# Data Frame
count_lang = tmdb.original_language.value_counts().to_frame().reset_index()
count_lang.head()
```

```python
sns.barplot(x="original_language",y="count",data = count_lang)
```

```python
sns.catplot(x="original_language", kind="count", data = tmdb)
```

```python
# Cuidado !!
# Grafico de pizza pode ser dificil de analise
plt.pie(count_lang["count"], labels= count_lang["original_language"])
```

```python
count_lang = tmdb.original_language.value_counts()
total_geral = count_lang.sum()
total_de_en = count_lang.loc["en"]
total_de_resto = total_geral - total_de_en
print(total_de_en,total_de_resto)
```

```python
dados = {
    'lingua': ['ingles', 'outros'],
    'total': [total_de_en, total_de_resto]
}
dados = pd.DataFrame(dados)
```

```python
sns.barplot(x='lingua', y='total', data=dados)
```

Buscar filmes diferente do ingles

```python
filmes_sem_lingua_original_ingles = tmdb.query("original_language != 'en'")
filmes_sem_lingua_original_ingles_count = filmes_sem_lingua_original_ingles.original_language.value_counts()
filmes_sem_lingua_original_ingles

```

```python
sns.catplot(x="original_language", kind="count", data=filmes_sem_lingua_original_ingles)
```

# Refinando visualizações

```python
sns.catplot(x="original_language", kind="count", data=filmes_sem_lingua_original_ingles, height=5, aspect=2, order=filmes_sem_lingua_original_ingles_count.index,palette="light:#5A9",hue='original_language')
```
